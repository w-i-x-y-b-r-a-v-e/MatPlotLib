```python
import matplotlib.pyplot as plt
import numpy as np

Geomatics = ("Pablo", "Sterling", "Nulu")
Gender = {
    "Male": np.array([10, 20, 5]),
    "Female":np.array([4, 12, 35])
}
width = 0.4

fig, ax = plt.subplots()
bottom = np.zeros(3)

for Gen, Gender, in Gender.items():
    Plot = ax.bar(Geomatics, Gender, width, label = Gender)
    ax.bar_label(Plot)
plt.legend()
plt.show()
```

```python
State = ["Lagos", "Anambra", "Imo", "Enugu"]
Abuja = [234, 167, 189, 130]
fig, ax = plt.subplots()
bar_con = ax.bar(State, Abuja)
ax.set(ylabel = "Distances km/hr", title = "Distance from Abuja to Differences States ", ylim = (0,260))
ax.bar_label(bar_con, fmt = lambda x:f' {x * 1.6:.2f} km/hr')
```

```python
np.random.seed(1234567)
Geomatics = ("Pablo", "Sterling", "Nulu", "collins")
y_pos = np.arange(len(Geomatics))
height = 3 + 20 * np.random.rand(len(Geomatics))
error = np.random.rand(len(Geomatics))

fig,ax = plt.subplots()
hbar = ax.barh(y_pos, height, xerr = error, align = "center")
ax.set_yticks(y_pos, labels = Geomatics)
ax.invert_yaxis()
ax.set_title("Number of Gender in Geomatics and Surveying")
ax.set_xlabel("values")
ax.set_ylabel("Name")
ax.bar_label(hbar, fmt = "%.2f")
ax.set_xlim(right = 12)
plt.show()
```

```python
np.random.seed(1234567)
Geomatics = ("Pablo", "Sterling", "Nulu", "collins")
y_pos = np.arange(len(Geomatics))
height = 3 + 20 * np.random.rand(len(Geomatics))
error = np.random.rand(len(Geomatics))

fig,ax = plt.subplots()
hbar = ax.barh(y_pos, height, xerr = error, align = "center")
ax.set_yticks(y_pos, labels = Geomatics)
ax.invert_yaxis()
ax.set_title("Number of Gender in Geomatics and Surveying")
ax.set_xlabel("values")
ax.set_ylabel("Name")
ax.bar_label(hbar, labels = [f' + {e:.2f}' for e in error],
            padding = 6, color = "r", fontsize = 10)
ax.set_xlim(right = 12)
plt.show()
```
